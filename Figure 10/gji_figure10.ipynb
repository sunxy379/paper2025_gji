```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib
%config InlineBackend.figure_format = 'retina'
```

```python
data_s01 = pd.read_excel('S-wave velocity model for python.xlsx',sheet_name = 'soils(calos)', header = 0)
data_s14 = pd.read_excel('S-wave velocity model for python.xlsx',sheet_name = 'soils(cha)', header = 0)
data_r = pd.read_excel('S-wave velocity model for python.xlsx',sheet_name = 'joint rocks', header = 0)
```

```python
fig, ax = plt.subplots(figsize = (5.86,3.66))

ax.scatter([12.45],[0.44],c = 'w', edgecolor = 'k',marker = 's', zorder = 7, label = 'Results for Figure 11')
ax.scatter([55.17,52.53,48.08,38.53],[0.35,0.36,0.39,0.42],c = 'w', edgecolor = 'tab:orange',zorder = 7, label = 'Results in this study')
ax.plot(data_s01['Theta'], data_s01['beta'],c = 'tab:pink',ls = 'dotted',lw = 2,label = 'Santamarina et al. (2001) for soils')
ax.plot(data_s14['Theta'], data_s14['beta'],c = 'tab:blue',ls = '-',lw = 2,label = 'Cha et al. (2014) for soils')
ax.plot(data_r['Theta'], data_r['beta'],c = 'tab:green',ls = '-',lw = 2,label = 'Cha et al. (2014) for jointed rocks')
ax.plot([5,80],[0.28,0.28], color="k", linestyle="--")
ax.plot([5,200],[0.16,0.16], color="k", linestyle="--")



ax.annotate('', xy=(15, 0.75-0.27*np.log10(15)), xytext=(30, 0.75-0.27*np.log10(30)),
            arrowprops=dict(arrowstyle = 'simple',facecolor='tab:gray', edgecolor = 'tab:gray'),
            )
ax.text(8.5,0.45,'Decreasing density\n         and plasticity',fontsize = 9,zorder = 8)

ax.annotate('', xy=(170, 0.71-0.27*np.log10(170)), xytext=(80, 0.71-0.27*np.log10(80)),
            arrowprops=dict(arrowstyle = 'simple',facecolor='tab:gray', edgecolor = 'tab:gray'),
            )
ax.text(56,0.08,'    Increasing\ncementation',fontsize = 9)

ax.grid('--')
ax.legend(framealpha=0,fontsize=9)
ax.set_ylim(0,0.6)
ax.set_xlabel(r"$\theta$-factor (m/s)")
ax.set_ylabel(r"$\beta$-exponent ()")

ax.text(5.5,0.31,'Soft clays')
ax.text(5.5,0.175,'Loose sands\n\nDense sands')
ax.text(5.5,0.032,'Stiff clays\n\nCemented soils')

ax.text(150,0.175,'Jointed rocks', rotation=-35.2)

ax.set_xlim(5, 800)
ax.set_xscale('log')
ax.set_xticks([5, 50, 500])
ax.get_xaxis().set_major_formatter(plt.ScalarFormatter()) 

# plt.savefig('power_law_figure_1117.png', dpi=600,bbox_inches = 'tight')

plt.show()
```
